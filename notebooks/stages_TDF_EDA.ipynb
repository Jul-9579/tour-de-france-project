```python
# Plot winning countries per year
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='year', hue='winner_country', palette='Set2')
plt.title('Winning Countries per Year')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Number of Wins')
plt.legend(title='Winner Country', bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.tight_layout()
plt.show()
```

```python
# Plot winning countries per year for the top 10 countries
top_countries = df['winner_country'].value_counts().nlargest(10).index  
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['winner_country'].isin(top_countries)], x='year', hue='winner_country', palette='Set2')
plt.title('Top 10 Winning Countries per Year')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Number of Wins')
plt.legend(title='Winner Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot winning countries per decade for the top 10 countries 
df['decade'] = (df['year'].astype(int) // 10) * 10
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['winner_country'].isin(top_countries)], x='decade', hue='winner_country', palette='Set2')
plt.title('Top 10 Winning Countries per Decade')
plt.xticks(rotation=45)
plt.xlabel('Decade')
plt.ylabel('Number of Wins')
plt.legend(title='Winner Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot winning countries per decade for the top 5 countries 
top_5_countries = df['winner_country'].value_counts().nlargest(5).index
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['winner_country'].isin(top_5_countries)], x='decade', hue='winner_country', palette='Set2')
plt.title('Top 5 Winning Countries per Decade')
plt.xticks(rotation=45)
plt.xlabel('Decade')
plt.ylabel('Number of Wins')
plt.legend(title='Winner Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Normalize the 'type' column
df['type'] = df['type'].str.lower().str.strip()
# Count occurrences of each unique value in 'type' column
type_counts = df['type'].value_counts()
type_counts.head(20)
```

```python
# Average distance per year
avg_distance_per_year = df.groupby('year')['distance'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_distance_per_year, x='year', y='distance', hue='year', palette='viridis', legend=False)
plt.title('Average Distance per Year')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Average Distance (km)')
# Show only every five years on the x-axis
years = avg_distance_per_year['year'].astype(int)
tick_positions = range(0, len(years), 5)
plt.xticks(tick_positions, years.iloc[tick_positions].astype(str), rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average distance per stage type
avg_distance_per_type = df.groupby('type')['distance'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(
    data=avg_distance_per_type.sort_values('distance', ascending=False),
    x='type',
    y='distance',
    hue='type',
    palette='viridis'  # hue removed
)

plt.title('Average Distance per Stage Type')
plt.xlabel('Stage Type (sorted by average distance)')
plt.ylabel('Average Distance (km)')
plt.xticks(rotation=45, ha='right')  # Rotate and align labels
plt.tight_layout()
plt.show()
```

```python
# Average distance per stage type per decade
avg_distance_per_type_decade = df.groupby(['decade', 'type'])['distance'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(
    data=avg_distance_per_type_decade,
    x='decade',
    y='distance',
    hue='type',
    palette='viridis'
)
plt.title('Average Distance per Stage Type per Decade')
plt.xlabel('Decade')
plt.ylabel('Average Distance (km)')
plt.xticks(rotation=45)
plt.legend(title='Stage Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Average distance per stage type per decade for the top 5 stage types from 1950 onwards
top_5_types = df[df['year'].astype(int) >= 1950]['type'].value_counts().nlargest(5).index
avg_distance_per_type_decade_top5 = df[df['type'].isin(top_5_types)].groupby(['decade', 'type'])['distance'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(
    data=avg_distance_per_type_decade_top5,
    x='decade',
    y='distance',
    hue='type',
    palette='pastel'  # Using a pastel color palette for better visibility
)
plt.title('Average Distance per Top 5 Stage Types per Decade (from 1950 onwards)')
plt.xlabel('Decade')
plt.ylabel('Average Distance (km)')
plt.xticks(rotation=45)
plt.legend(title='Stage Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Group stage types by category
stage_type_mapping = {
    'flat stage': 'Flat',
    'plain stage': 'Flat',
    'flat cobblestone stage': 'Flat',
    'plain stage with cobblestones': 'Flat',
    'transition stage': 'Flat',
    'half stage': 'Flat',  # Optional: treat specially if mixed

    'medium mountain stage': 'Hilly',
    'hilly stage': 'Hilly',
    'intermediate stage': 'Hilly',

    'high mountain stage': 'Mountain',
    'mountain stage': 'Mountain',
    'stage with mountain(s)': 'Mountain',
    'stage with mountain': 'Mountain',

    'individual time trial': 'Time Trial',
    'team time trial': 'Time Trial',
    'mountain time trial': 'Time Trial',
}
```

```python
df['stage_category'] = df['type'].map(stage_type_mapping)
```

```python
# Ensure 'year' is of integer type if needed
df['year'] = df['year'].astype(int)

# Filter for 'Plain stage' and count per year
plain_stage_counts = (
    df[df['type'] == 'Plain stage']
    .groupby('year')
    .size()
    .reset_index(name='plain_stage_count')
)

# Preview the result

plain_stage_counts
```

```python
# Add average distance of plain stages per year to the plain_stage_counts table
plain_stage_counts = plain_stage_counts.merge(
    df[df['type'] == 'Plain stage'].groupby('year')['distance'].mean().reset_index(name='avg_plain_stage_distance'),
    on='year',
    how='left'
)

plain_stage_counts
```

```python
# Ensure year is numeric
df['year'] = df['year'].astype(int)

# Filter for 'Flat stage' and group by year to get the average distance
avg_flat_stage_distance = (
    df[df['type'] == 'Flat stage']
    .groupby('year')['distance']
    .mean()
    .reset_index(name='avg_distance')
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

plt.figure(figsize=(14, 6))
sns.barplot(
    data=avg_flat_stage_distance,
    x='year',
    y='avg_distance',
    color='royalblue'
)

plt.title('Average Distance of Flat Stages Per Year (1950s Onwards)')
plt.xlabel('Year')
plt.ylabel('Average Distance (km)')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
